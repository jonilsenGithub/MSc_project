# Euclidean Distance

```python
import numpy as np
import pandas as pd
```

```python
#importing dataset
org_df = pd.read_csv('keystroke.csv')
org_df
```

```python
#formatting subject column to ints
org_df['subject'] = org_df['subject'].apply(lambda x: int(x[2:]))
```

```python
#splitting half the samples into trainingset and the other half to testingset, based on sessionIndex column
training_set = org_df[(org_df['sessionIndex']==1) | (org_df['sessionIndex']==3) | (org_df['sessionIndex']==5) | (org_df['sessionIndex']==7)]
test_set = org_df[(org_df['sessionIndex']==2) | (org_df['sessionIndex']==4) | (org_df['sessionIndex']==6) | (org_df['sessionIndex']==8)]
```

```python
#removing the obsolete columns sessionIndex and rep
training_set = training_set.drop(['sessionIndex','rep'],axis=1)
test_set = test_set.drop(['sessionIndex','rep'],axis=1)
```

```python
#instanciating scaler
from sklearn.preprocessing import QuantileTransformer
scaler = QuantileTransformer(output_distribution='normal')

#fitting scaler to training_set, but transforming both traning_set and test_set
training_set[['H.period', 'DD.period.t', 'UD.period.t', 'H.t', 'DD.t.i',
       'UD.t.i', 'H.i', 'DD.i.e', 'UD.i.e', 'H.e', 'DD.e.five', 'UD.e.five',
       'H.five', 'DD.five.Shift.r', 'UD.five.Shift.r', 'H.Shift.r',
       'DD.Shift.r.o', 'UD.Shift.r.o', 'H.o', 'DD.o.a', 'UD.o.a', 'H.a',
       'DD.a.n', 'UD.a.n', 'H.n', 'DD.n.l', 'UD.n.l', 'H.l', 'DD.l.Return',
       'UD.l.Return', 'H.Return']] = scaler.fit_transform(training_set[['H.period', 'DD.period.t', 'UD.period.t', 'H.t', 'DD.t.i',
       'UD.t.i', 'H.i', 'DD.i.e', 'UD.i.e', 'H.e', 'DD.e.five', 'UD.e.five',
       'H.five', 'DD.five.Shift.r', 'UD.five.Shift.r', 'H.Shift.r',
       'DD.Shift.r.o', 'UD.Shift.r.o', 'H.o', 'DD.o.a', 'UD.o.a', 'H.a',
       'DD.a.n', 'UD.a.n', 'H.n', 'DD.n.l', 'UD.n.l', 'H.l', 'DD.l.Return',
       'UD.l.Return', 'H.Return']])

test_set[['H.period', 'DD.period.t', 'UD.period.t', 'H.t', 'DD.t.i',
       'UD.t.i', 'H.i', 'DD.i.e', 'UD.i.e', 'H.e', 'DD.e.five', 'UD.e.five',
       'H.five', 'DD.five.Shift.r', 'UD.five.Shift.r', 'H.Shift.r',
       'DD.Shift.r.o', 'UD.Shift.r.o', 'H.o', 'DD.o.a', 'UD.o.a', 'H.a',
       'DD.a.n', 'UD.a.n', 'H.n', 'DD.n.l', 'UD.n.l', 'H.l', 'DD.l.Return',
       'UD.l.Return', 'H.Return']] = scaler.transform(test_set[['H.period', 'DD.period.t', 'UD.period.t', 'H.t', 'DD.t.i',
       'UD.t.i', 'H.i', 'DD.i.e', 'UD.i.e', 'H.e', 'DD.e.five', 'UD.e.five',
       'H.five', 'DD.five.Shift.r', 'UD.five.Shift.r', 'H.Shift.r',
       'DD.Shift.r.o', 'UD.Shift.r.o', 'H.o', 'DD.o.a', 'UD.o.a', 'H.a',
       'DD.a.n', 'UD.a.n', 'H.n', 'DD.n.l', 'UD.n.l', 'H.l', 'DD.l.Return',
       'UD.l.Return', 'H.Return']])
```

```python
#grouping samples by username (subject), extracting feature means, re-indexing dataset 
# and saving it to the variable df
df = training_set[['subject','H.period', 'DD.period.t', 'UD.period.t', 'H.t', 'DD.t.i',
       'UD.t.i', 'H.i', 'DD.i.e', 'UD.i.e', 'H.e', 'DD.e.five', 'UD.e.five',
       'H.five', 'DD.five.Shift.r', 'UD.five.Shift.r', 'H.Shift.r',
       'DD.Shift.r.o', 'UD.Shift.r.o', 'H.o', 'DD.o.a', 'UD.o.a', 'H.a',
       'DD.a.n', 'UD.a.n', 'H.n', 'DD.n.l', 'UD.n.l', 'H.l', 'DD.l.Return',
       'UD.l.Return', 'H.Return']].groupby('subject').mean().reset_index()

df_no_users = df.drop('subject',axis=1)
```

```python
from sklearn.metrics.pairwise import euclidean_distances

#this variable keeps track of what is currently the best euclidean distance anomaly score
best_score = 0

#this variable keeps track of what user the current best_score belongs to
pred_user = 0

#this variable counts number of correct predictions
count = 0

#for-loop iterating through test_set
for row in range(0,10200):
    
    #processing data from request by formatting and separating
    # username and measurements into the variables test_profile
    # and test_subject
    test = np.array(test_set.iloc[row]).reshape(1,-1)
    test_profile = np.delete(test,0,axis=1)
    test_subject = test[0][0]

    #for-loop iterating through all user profiles
    for i in range(0,51):

        #formatting the profile to use it in euclidean_distances() method
        training = df_no_users.iloc[i]
        training_profile = np.array(training).reshape(1, -1)
        
        #an IF statement storing the first profile as best_score and pred_user
        if i == 0:
            best_score = euclidean_distances(test_profile,training_profile)
            pred_user = df['subject'].iloc[i]
            
        #an ELIF statement only storing the next profile as best_score and pred_user if 
        # the euclidean_distance between test_profile and training_profile is less than the current best_score
        elif i != 0 and euclidean_distances(test_profile,training_profile) < best_score:

            best_score = euclidean_distances(test_profile,training_profile)
            pred_user = df['subject'].iloc[i]
           
        #if none of the above is true then best_score and pred_user remains the same
        else:
            best_score = best_score
            pred_user = pred_user
    
    #an IF statement incrementing the count variable only the user prediction matches the actual sample's
    # user
    if test_subject==pred_user: 
        
        count = count+1
    
#print statement printing accuracy score
print('Accuracy rate is: {}'.format(count/10200))
```
