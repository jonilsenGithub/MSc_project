## Type 1 error testing

### One sample two-sided t-test

```python
from scipy.stats import ttest_1samp

count = 0

for row in range(0,1000):
    

    user_input = np.array(org_df.iloc[random.randint(0,18488)]).reshape(1,-1)
    x = np.delete(user_input,0,axis=1)
    y = user_input[0][0]

    sample = np.array(x[0])

    profile = np.array(df_no_users.iloc[int(df[df['subject']==y].index.values)])
    
    ttest, pval = ttest_1samp(sample,profile.mean())

    if pval > 0.05:
        count = count+1

print('Accepts true null hypothesis: {}'.format(count/1000))
```

### Independent t-test

```python
from scipy.stats import ttest_ind

count = 0

for row in range(0,1000):
    

    user_input = np.array(org_df.iloc[random.randint(0,18487)]).reshape(1,-1)
    x = np.delete(user_input,0,axis=1).reshape(1,-1)
    y = user_input[0][0]

    sample = x[0]

    profile = np.array(df_no_users.iloc[int(df[df['subject']==y].index.values)])
    
    ttest, pval = ttest_ind(sample,profile)
    
    if pval > 0.05:
        count = count+1

print('Accepts true null hypothesis: {}'.format(count/1000))
```

### Mann-Whitney test

```python
from scipy.stats import mannwhitneyu

count = 0

for row in range(0,1000):
    

    user_input = np.array(org_df.iloc[random.randint(0,18487)]).reshape(1,-1)
    x = np.delete(user_input,0,axis=1)
    y = user_input[0][0]

    sample = np.array(x[0])

    profile = np.array(df_no_users.iloc[int(df[df['subject']==y].index.values)])
    
    ttest, pval = mannwhitneyu(sample,profile)

    if pval > 0.05:
        count = count+1

print('Accepts true null hypothesis: {}'.format(count/1000))
```

## Type 2 error testing

### One sample two-sided t-test

```python
from scipy.stats import ttest_rel

count=0

for row in range(0,1000):
    
    fraud_usr = org_df.iloc[random.randint(0,18488)]
    fraud_sub = fraud_usr['subject']
    fraud_usr_data = fraud_usr.drop('subject')
    sample = np.array(fraud_usr_data)
    
    index = int(df[df['subject']==int(fraud_sub)].index.values)
    
    if len(range(0,index)) > len(range(index,50)):
        attempt = random.randint(0,index-1)
    else:
        attempt = random.randint(index+1,50)
    
    ac_usr = df.iloc[attempt]
    ac_sub = ac_usr['subject']
    ac_usr_data = ac_usr.drop('subject')
    profile = np.array(ac_usr_data)
    
    ttest,pval = ttest_rel(sample,profile)
    
    if pval < 0.05:
        count=count+1

print('Rejects false null hypothesis: {}'.format(count/1000))
```
