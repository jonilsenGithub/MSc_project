# Neural Network

```python
import numpy as np
import pandas as pd
from tensorflow import keras
```

```python
df = pd.read_csv('keystroke.csv')
```

```python
df = df.drop(['sessionIndex','rep'],axis=1)
```

```python
df['subject'] = df['subject'].apply(lambda x: int(x[2:]))
```

```python
df
```

## Reassigning subject values

```python
new_subject = {2:1, 3:2, 4:3, 5:4, 7:5, 8:6, 10:7, 11:8, 12:9, 13:10, 15:11, 16:12, 17:13, 18:14, 19:15, 20:16, 21:17,
       22:18, 24:19, 25:20, 26:21, 27:22, 28:23, 29:24, 30:25, 31:26, 32:27, 33:28, 34:29, 35:30, 36:31, 37:32, 38:33, 39:34,
       40:35, 41:36, 42:37, 43:38, 44:39, 46:40, 47:41, 48:42, 49:43, 50:44, 51:45, 52:46, 53:47, 54:48, 55:49, 56:50, 57:51}
```

```python
df['subject'] = df['subject'].replace(new_subject)
```

## Splitting data

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('subject',axis=1)
y = keras.utils.to_categorical(np.array(df['subject']).reshape(20400,1),52)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

## Preprocessing

```python
from sklearn.preprocessing import QuantileTransformer
```

```python
scaler = QuantileTransformer(output_distribution='normal')
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

## Sequential model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()

model.add(Dense(40, input_dim=31, activation='relu'))
model.add(Dense(40, activation='relu'))
model.add(Dense(52, activation='softmax'))

model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(x=X_train,y=y_train,batch_size=256,epochs=500,verbose=1)
```

```python
score = model.evaluate(X_test,y_test)
```
